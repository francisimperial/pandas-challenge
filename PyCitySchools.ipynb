```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data = school_data.rename(columns={'school_name':'School','type':'Type','size':'Size','budget':'Budget'})
student_data = student_data.rename(columns={'school_name':'School','student_name':'Name','gender':'Gender','grade':'Grade',
                                           'reading_score':'Reading Score','math_score':'Math Score'})
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["School", "School"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = school_data['School'].count()
total_students = school_data_complete['Student ID'].count()
total_budget = school_data['Budget'].sum()

avg_math = school_data_complete['Math Score'].mean()
avg_reading = school_data_complete['Reading Score'].mean()
overall_score = (avg_math + avg_reading)/2

math_percent = school_data_complete.loc[school_data_complete['Math Score'] >= 70]['Student ID'].count()/total_students * 100
reading_percent = school_data_complete.loc[school_data_complete['Reading Score'] >= 70]['Student ID'].count()/total_students * 100

#I was told to find the weighted average for the overall passing rate
#However I don't understand how this would find the overall passing rate when it's larger than the % Passing Math
#This method also still doesn't match what is on the solution so I'm not sure if this even is 'correct'
overall_count = school_data_complete.loc[school_data_complete['Math Score'] >= 70]['Student ID'].count() + school_data_complete.loc[school_data_complete['Reading Score'] >= 70]['Student ID'].count()
math_count = school_data_complete.loc[school_data_complete['Math Score'] >= 70]['Student ID'].count()
reading_count = school_data_complete.loc[school_data_complete['Reading Score'] >= 70]['Student ID'].count()
overall_percent = (math_count/overall_count) * math_percent + (reading_count/overall_count) * reading_percent

#This is how I would convey the Overall Passing Rate
#I continue to include this in the following codes because I just want to show a comparison 
overall_percent_actual = school_data_complete.loc[(school_data_complete['Math Score'] >= 70) & (school_data_complete['Reading Score'] >= 70)]['Student ID'].count()/total_students * 100

district_summary = pd.DataFrame({
    'Total Schools':[total_schools],
    'Total Students':[total_students],
    'Total Budget':[total_budget],
    'Average Math Score':[avg_math],
    'Average Reading Score':[avg_reading],
    'Overall Average Score':[overall_score],
    '% Passing Math':[math_percent],
    '% Passing Reading':[reading_percent],
    '% Overall Passing Rate':[overall_percent],
    'Actual Overall Passing Rate':[overall_percent_actual]
})

#reorder columns
dist_summary = district_summary[[
    'Total Schools',
    'Total Students',
    'Total Budget',
    'Average Math Score',
    'Average Reading Score',
    'Overall Average Score',
    '% Passing Math',
    '% Passing Reading',
    '% Overall Passing Rate',
    'Actual Overall Passing Rate'
]]

#formatting
dist_summary.style.format({
    'Total Students':'{:,}',
    'Total Budget':'${:,.2f}'
})
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
group_schools = school_data_complete.set_index('School').groupby(['School'])

school_type = school_data.set_index('School')['Type']
total_students = school_data.set_index('School')['Size']
total_budget = school_data.set_index('School')['Budget']

per_student_budget = total_budget/total_students

avg_math = group_schools['Math Score'].mean()
avg_reading = group_schools['Reading Score'].mean()

math_percent = school_data_complete[school_data_complete['Math Score'] >= 70].groupby('School')['Student ID'].count()/total_students * 100
reading_percent = school_data_complete[school_data_complete['Reading Score'] >= 70].groupby('School')['Student ID'].count()/total_students * 100
overall_count = school_data_complete.loc[school_data_complete['Math Score'] >= 70]['Student ID'].count() + school_data_complete.loc[school_data_complete['Reading Score'] >= 70]['Student ID'].count()
math_count = school_data_complete.loc[school_data_complete['Math Score'] >= 70]['Student ID'].count()
reading_count = school_data_complete.loc[school_data_complete['Reading Score'] >= 70]['Student ID'].count()
overall_percent = (math_count/overall_count) * math_percent + (reading_count/overall_count) * reading_percent

overall_percent_actual = school_data_complete[(school_data_complete['Math Score'] >= 70) & (school_data_complete['Reading Score'] >= 70)].groupby('School')['Student ID'].count()/total_students * 100

school_summary = pd.DataFrame({
    'School Type':school_type,
    'Total Students':total_students,
    'Total School Budget':total_budget,
    'Per Student Budget':per_student_budget,
    'Average Math Score':avg_math,
    'Average Reading Score':avg_reading,
    '% Passing Math':math_percent,
    '% Passing Reading':reading_percent,
    '% Overall Passing Rate':overall_percent,
    'Actual Overall Passing Rate':overall_percent_actual
})

#reorder columns
school_sum = school_summary[[
    'School Type',
    'Total Students',
    'Total School Budget',
    'Per Student Budget',
    'Average Math Score',
    'Average Reading Score',
    '% Passing Math',
    '% Passing Reading',
    '% Overall Passing Rate',
    'Actual Overall Passing Rate'
]]

#formatting
school_sum.style.format({
    'Total Students':'{:,}',
    'Total School Budget':'${:,.2f}',
    'Per Student Budget':'${:,.2f}'
})
```

## Top Performing Schools (By Passing Rate)

```python
top_five = school_sum.sort_values('% Overall Passing Rate', ascending = False)

#format and show top five schools
top_five.head(5).style.format({
    'Total Students':'{:,}',
    'Total School Budget':'${:,.2f}',
    'Per Student Budget':'${:,.2f}'
})
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
school_data_complete['Spending Category'] = pd.cut(school_data_complete['Budget']/school_data_complete['Size'],spending_bins,labels=group_names)
group_spending = school_data_complete.groupby('Spending Category')

avg_math = group_spending['Math Score'].mean()
avg_reading = group_spending['Reading Score'].mean()

math_percent = school_data_complete[school_data_complete['Math Score']>=70].groupby('Spending Category')['Student ID'].count()/group_spending['Student ID'].count() * 100
reading_percent = school_data_complete[school_data_complete['Reading Score']>=70].groupby('Spending Category')['Student ID'].count()/group_spending['Student ID'].count() * 100
overall_count = school_data_complete.loc[school_data_complete['Math Score'] >= 70].groupby('Spending Category')['Student ID'].count() + school_data_complete.loc[school_data_complete['Reading Score'] >= 70].groupby('Spending Category')['Student ID'].count()
math_count = school_data_complete.loc[school_data_complete['Math Score'] >= 70].groupby('Spending Category')['Student ID'].count()
reading_count = school_data_complete.loc[school_data_complete['Reading Score'] >= 70].groupby('Spending Category')['Student ID'].count()
overall_percent = (math_count/overall_count) * math_percent + (reading_count/overall_count) * reading_percent

overall_percent_actual = school_data_complete[(school_data_complete['Math Score']>=70) & (school_data_complete['Reading Score']>=70)].groupby('Spending Category')['Student ID'].count()/group_spending['Student ID'].count() * 100

spending_summary = pd.DataFrame({
    'Average Math Score':avg_math,
    'Average Reading Score':avg_reading,
    '% Passing Math':math_percent,
    '% Passing Reading':reading_percent,
    '% Overall Passing Rate':overall_percent,
    'Actual Overall Passing Rate':overall_percent_actual
})

spending_summary
```
